```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
start=time.time()
```

```python
df=pd.read_csv("bank-additional-full.csv",sep=';')
```

```python
numeric_columns=list(df._get_numeric_data().columns)
categorical_columns=list(set(df.columns)-set(numeric_columns)-set('y'))
categorical_columns
```

# Barplot For Categorical Variable

```python
fig , axis = plt.subplots(2, 5, figsize=(25,25))
fig.suptitle('barplots of categorical variable')
i=0
for x in categorical_columns:
    p=sns.countplot(ax=axis[i//5,i%5],x=list(df[x]),data=df)
    p.set_title(x)
    p.set_xticklabels(p.get_xticklabels(), rotation=90)
    i=i+1
plt.show()
```

# Histograms of Numerical Variable

```python
fig , axis = plt.subplots(2, 5, figsize=(20, 10))
fig.suptitle('histograms of numerical variable')
i=0
for x in numeric_columns:
    axis[i//5,i%5].hist(df[x],color='green')
    axis[i//5,i%5].set_title(x)
    i=i+1
plt.show() 
```

# Barplot of Target Variable

```python
y=list(df['y'])
a={i:y.count(i) for i in y}
plt.bar(list(a.keys()),list(a.values()),color='blue')
plt.show()
```
